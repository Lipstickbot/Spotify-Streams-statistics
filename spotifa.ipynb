```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv(r'C:\Users\vovot\Downloads\spotif\spoti2024.csv', encoding='latin1')

# Ensure 'Spotify Streams' column is numeric
df['Spotify Streams'] = pd.to_numeric(df['Spotify Streams'], errors='coerce')

# Sort the DataFrame
sorted_df = df.sort_values(by='Spotify Streams', ascending=False)
# Plot the data
sorted_df.plot(kind='bar')
plt.ylabel('Spotify Streams')
plt.title('Spotify Streams for Selected Rows')
plt.show()

```
